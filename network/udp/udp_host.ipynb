## udp server sample

bind 인자로는 아이피주소와 포트가 들어간간다. bind(ip,port) ip는 특별히 지정하지않으려면 '' 넣어줄수있다.
지정된 포트값에 의하여 해당 포트를 열고 패킷이 들어오기를 기다린다.

recvform 함수는 패킷이 날라 올때까지 대기하는 블럭킹함수이다. 이함수에 인자는 최대 한번에 받을수있는 바이트 수를 지정한다.


```python
import socket
from struct import *

udp_socket = socket.socket(
    socket.AF_INET, #internet
    socket.SOCK_DGRAM # udp
    )
udp_socket.bind(('',8284))

print('init socket')

while True:
    print('wait packet...')
    _data, _rinfo = udp_socket.recvfrom(1024) # buffer size is 1024 bytes
    _packet = unpack("<LBBhf",_data)
    print(_packet)
    if _packet[0] == 99 :
        _res = pack('<LBBHH',77,0,9,300,400)
        udp_socket.sendto(_res,(_rinfo[0],_rinfo[1]))
    else :
        _res = pack('<LBBHH',77,0,0,0,0)
    
```
